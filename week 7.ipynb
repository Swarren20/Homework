```python
import numpy as np

#1

A = np.array([[25, 16]])
B = np.array([[11, 2], [13, 4]])
C = np.array([[7, 81], [5, 6], [11, 12]])

#2

newArray = np.concatenate((A, B, C))
print(newArray)

#3

sortedArray = np.sort(newArray)
print("\n", sortedArray)

#4

reshapedArray = sortedArray.reshape(2,2,3)
print("\n", reshapedArray)

#5

import numpy as np


d, e = reshapedArray

a0, a1 = d
b0, b1 = e
print("\nUnpacked tuple into 4 variables...", a0)
print(a1)
print(b0)
print(b1)

#6

one = np.array([10, 11, 12, 13, 14, 15, 16, 17])
two = np.array([20, 21, 22, 23, 24, 25, 26, 27])
three = np.array([ 0, 1, 2, 3, 4, 5, 6, 7])

comboArray = np.concatenate(((one, two, three)))
print(comboArray)

print("Number of Elements...", len(comboArray))
sec1 = comboArray[2]
print("sec1...", sec1)
sec2 = comboArray[3:24]
print("sec2...", sec2)
sec3 = comboArray[4:15]
print("sec3...", sec3)
sec4 = comboArray[-6:]
print("sec4...", sec4)
sec5 = comboArray[:-8]
print("sec5...", sec5)
sec6 = comboArray[20:25:2]
print("sec6...", sec6)
sec7 = comboArray[::-5]
print("sec7...", sec7)
```

```python

#7

import numpy as np

Array1 = np.array([[[ 0.0,  0.0,  0.10], [10.0, 10.30, 10.01]],
        [[20.0, 20.02, 20.10], [30.0, 31.0, 30.30]],
        [[40.40, 24.0, 40.0], [15.0, 15.5, 25.5]]])

Array2 = np.array([[17.0, 70.0, 10.7]])

Addition = np.add(Array1, Array2)
print("Addition: \n", Addition)

Division = np.divide(Array1, Array2)
print("\n Division: \n", Division)

Times = np.multiply(Array1, Array2)
print("\n Multiplication: \n", Times)


#8

Subtraction = np.subtract(Array1, Array2)
print("\n Subtraction: \n", Subtraction)

#9 

Divisible1 = Array1[Array1 % 3 == 0]
print("\n Divisible by 3 for Array 1: \n", Divisible1)




```

```python
#10

import pandas as pd

Ingredients = pd.Series(["Flour", "Milk", "Eggs", "Spam"], name='Ingredients')
Quantity = pd.Series([4, 1, 2, 1], name='Quantity')
Unit = pd.Series(["cups", "cup", "large", "can"], name='Unit')

index_labels = ["Item 1", "Item 2", "Item 3", "Item 4"]
Ingredients.index = index_labels
Quantity.index = index_labels
Unit.index = index_labels

Dinner = pd.concat([Ingredients,Quantity,Unit], axis=1)
print(Dinner)
```

```python
import pandas as pd

# 11

data = {'Name': ['Jai', 'janusha', 'Gaurav', 'Anuj'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
        'address': ['Delhi', 'Doha', 'Chennai', 'Dakhar'],
        'Age': [21, 23, 24, 21],
        'Pets': ['Dog', 'Bunny', 'Chinchilla', 'Parrot'],
        'sport': ['Darts', 'Basketball', 'PaddleBoarding', 'Cricket']
    }

studentData = pd.DataFrame(data)
print(studentData)

#12

studentData.insert(7, "Desserts", ["ice cream", "Cashew Fudge", "waffels", "Carrot Halwa"], True)
print("\n", studentData)

#13

studentData.sort_values(by = ['Name'], inplace = True)
print("\n Sorted by Name: \n", studentData)

studentData.sort_values(by = ['address'], inplace = True)
print("\n Sorted by Address: \n", studentData)
```
